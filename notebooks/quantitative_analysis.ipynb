```python
import pandas as pd
import talib
import matplotlib.pyplot as plt
import os
```

```python
# Define the list of  data files ---

data_files = [
    'AAPL_historical_data.csv',
    'AMZN_historical_data.csv',
    'GOOG_historical_data.csv',
    'META_historical_data.csv',
    'MSFT_historical_data.csv',
    'NVDA_historical_data.csv',
    'TSLA_historical_data.csv'
]
```

```python
# Define the path to  data directory
data_dir = 'data/' # Updated data directory path
```

```python
# --- Loop through each data file ---
for file_name in data_files:
    file_path = os.path.join(data_dir, file_name)

    print(f"\n--- Processing data from: {file_path} ---")

    try:
       
        df = pd.read_csv(file_path, parse_dates=['Date'], index_col='Date')

       
        df.columns = [col.capitalize() for col in df.columns]
        required_cols = ['Open', 'High', 'Low', 'Close', 'Volume']
        if not all(col in df.columns for col in required_cols):
            print(f"Warning: Not all required OHLCV columns found in {file_name}.")
            print(f"Expected: {required_cols}, Found: {df.columns.tolist()}")
           
            column_mapping = {
                'Adj Close': 'Close', 
                'Adjclose': 'Close',
                'Datetime': 'Date'
            }
            df.rename(columns=column_mapping, inplace=True)
            if not all(col in df.columns for col in required_cols):
                print("Attempted column mapping, but still missing required OHLCV columns. Please check your CSV column names.")
                #  let's create dummy columns if they are missing
                for col in required_cols:
                    if col not in df.columns:
                        print(f"Creating dummy column for missing: {col}")
                        df[col] = df['Close'] 
                if 'Volume' not in df.columns:
                    df['Volume'] = 1000000 

        # Sort by date to ensure correct calculation for time-series indicators
        df.sort_index(inplace=True)

        print(f"Data loaded successfully from {file_name}. First 5 rows:")
        print(df.head())
        print(f"DataFrame shape: {df.shape}")

    except FileNotFoundError:
        print(f"Error: The file '{file_path}' was not found. Skipping this file.")
        continue # Skip to the next file in the list
    except KeyError as e:
        print(f"Error: Missing expected column in CSV '{file_name}': {e}. Please check your CSV headers. Skipping this file.")
        continue # Skip to the next file
    except Exception as e:
        print(f"An unexpected error occurred while loading '{file_name}': {e}. Skipping this file.")
        continue # Skip to the next file


    # --- 2. Apply Analysis Indicators with TA-Lib ---
    print(f"\nCalculating Technical Indicators with TA-Lib for {file_name}...")

    # Simple Moving Average (SMA)
    df['SMA_20'] = talib.SMA(df['Close'], timeperiod=20)
    df['SMA_50'] = talib.SMA(df['Close'], timeperiod=50)

    # Relative Strength Index (RSI)
    df['RSI'] = talib.RSI(df['Close'], timeperiod=14)

    # Moving Average Convergence Divergence (MACD)
    macd, macdsignal, macdhist = talib.MACD(df['Close'],
                                            fastperiod=12,
                                            slowperiod=26,
                                            signalperiod=9)
    df['MACD'] = macd
    df['MACD_Signal'] = macdsignal
    df['MACD_Hist'] = macdhist

    print("Technical indicators calculated. Last 5 rows with new indicators:")
    print(df.tail())


    # --- 3. Use PyNance for Financial Metrics ---
    try:
        print(f"\nUsing of PyNance for {file_name}...")
        df['Daily_Return'] = df['Close'].pct_change()
        df['Volatility'] = df['Daily_Return'].rolling(window=20).std() * (252**0.5)
        print("Basic financial metrics calculated (Daily Return, Volatility).")
        print(df[['Close', 'Daily_Return', 'Volatility']].tail())

    except ImportError:
        print("PyNance not found")
        print("Proceeding with basic financial metrics using pandas.")
        df['Daily_Return'] = df['Close'].pct_change()
        df['Volatility'] = df['Daily_Return'].rolling(window=20).std() * (252**0.5)
    except Exception as e:
        print(f"An error occurred with PyNance or financial metric calculation for {file_name}: {e}")
        df['Daily_Return'] = df['Close'].pct_change()
        df['Volatility'] = df['Daily_Return'].rolling(window=20).std() * (252**0.5)


    # --- 4. Visualize the Data ---
    print(f"\nGenerating visualizations for {file_name}...")

    # Plotting Close Price and Moving Averages
    plt.figure(figsize=(14, 7))
    plt.plot(df['Close'], label='Close Price', color='blue', alpha=0.7)
    plt.plot(df['SMA_20'], label='SMA 20', color='red', alpha=0.7)
    plt.plot(df['SMA_50'], label='SMA 50', color='green', alpha=0.7)
    plt.title(f'Stock Price with Moving Averages for {file_name}')
    plt.xlabel('Date')
    plt.ylabel('Price')
    plt.legend()
    plt.grid(True)
    plt.tight_layout()
    plt.show()

    # Plotting RSI
    plt.figure(figsize=(14, 4))
    plt.plot(df['RSI'], label='RSI', color='purple', alpha=0.7)
    plt.axhline(70, color='red', linestyle='--', label='Overbought (70)')
    plt.axhline(30, color='green', linestyle='--', label='Oversold (30)')
    plt.title(f'Relative Strength Index (RSI) for {file_name}')
    plt.xlabel('Date')
    plt.ylabel('RSI Value')
    plt.legend()
    plt.grid(True)
    plt.tight_layout()
    plt.show()

    # Plotting MACD
    plt.figure(figsize=(14, 6))
    plt.plot(df['MACD'], label='MACD', color='orange', alpha=0.7)
    plt.plot(df['MACD_Signal'], label='Signal Line', color='blue', alpha=0.7)
    plt.bar(df.index, df['MACD_Hist'], label='Histogram', color='gray', alpha=0.5)
    plt.axhline(0, color='black', linestyle='--')
    plt.title(f'MACD for {file_name}')
    plt.xlabel('Date')
    plt.ylabel('MACD Value')
    plt.legend()
    plt.grid(True)
    plt.tight_layout()
    plt.show()

    print(f"Visualizations generated successfully for {file_name}.")
```
