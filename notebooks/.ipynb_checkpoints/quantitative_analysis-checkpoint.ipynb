```python
import pandas as pd
import os
os.chdir(r'D:\10academy\financial-news-analysis')
import sys
sys.path.append(os.getcwd()) # Add the current working directory to Python's path
from scripts.data_loader import load_and_prepare_data
from scripts.indicator_calculation import calculate_technical_indicators, calculate_financial_metrics
from scripts.data_visualization import plot_stock_price_with_sma, plot_rsi, plot_macd, plot_volatility, plot_daily_return
import matplotlib.pyplot as plt 
import warnings
warnings.filterwarnings("ignore")
```

```python
# Define the list of  data files ---

data_files = [
    'AAPL_historical_data.csv',
    'AMZN_historical_data.csv',
    'GOOG_historical_data.csv',
    'META_historical_data.csv',
    'MSFT_historical_data.csv',
    'NVDA_historical_data.csv',
    'TSLA_historical_data.csv'
]
```

```python
# Define the path to  data directory
data_dir = 'notebooks/data/' 
```

```python
for file_name in data_files:
    file_path = os.path.join(data_dir, file_name)
    print(f"\n--- Processing data for: {file_name} ---")

    df = load_and_prepare_data(file_path)
    if df is not None:
        print(f"\nCalculating indicators for {file_name}...")
        df = calculate_technical_indicators(df)
        df = calculate_financial_metrics(df)
        print("Indicators and metrics calculated.")
        print(df[['Close', 'SMA_20', 'SMA_50', 'RSI', 'MACD', 'Volatility', 'Daily_Return']].tail())

        print(f"\nGenerating visualizations for {file_name}...")
        plot_stock_price_with_sma(df.copy(), file_name.replace('_historical_data.csv', ''))
        plot_rsi(df.copy(), file_name.replace('_historical_data.csv', ''))
        plot_macd(df.copy(), file_name.replace('_historical_data.csv', ''))
        plot_volatility(df.copy(), file_name.replace('_historical_data.csv', ''))
        plot_daily_return(df.copy(), file_name.replace('_historical_data.csv', ''))
print("\n--- All quantitative analysis tasks completed for all specified data files. ---")

print("\nRreview the generated plots and data.")
        
```
